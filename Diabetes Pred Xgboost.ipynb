```python
train_dataset = pd.read_csv('datasets_23663_30246_diabetes.csv')
X = train_dataset.iloc[:, :-1].values
y = train_dataset.iloc[:, -1].values
```

```python
df= pd.DataFrame(train_dataset)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier(random_state = 42,learning_rate = 0.1, colsample_bytree=0.5,reg_alpha=0.5, max_delta_step=1, min_child_weight=0.1)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
